```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the graph
years = ["Dec-2021", "Dec-2022", "Dec-2023", "Sept-2024"]
operating_cash_flow = [10845.8, 8486.8, 5443.4, 4818.4]
investing_cash_flow = [-72, -1028.9, -2689.3, -1993]
financing_cash_flow = [-9891.7, -7138.3, -3003.9, -2809.8]
free_cash_flow = [9905.5, 7167.5, 3247.2, 2839]

# Setting up the figure and axes
fig, ax = plt.subplots(figsize=(10, 5))
width = 0.25  # Bar width
x = np.arange(len(years))

# Plot bars
ax.bar(x - width, operating_cash_flow, width, label='Operating Cash Flow', color='#49c3fb')
ax.bar(x, investing_cash_flow, width, label='Investing Cash Flow', color='#00cadc')
ax.bar(x + width, financing_cash_flow, width, label='Financing Cash Flow', color='#0f238c')
ax.bar(x + 2 * width, free_cash_flow, width, label='Free Cash Flow', color='green')


# Formatting
ax.set_xticks(x)
ax.set_xticklabels(years)
ax.set_ylabel("(€ million)")
ax.set_title("ASML Cash Flow Statement (2021 - Sept 2024)")
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Display the graph
plt.show()
```
